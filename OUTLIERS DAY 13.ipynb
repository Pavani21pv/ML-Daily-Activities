```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "Study_Hours":[1,2,3,4,5,6,7,8,20]
})
    
```

data

# IQR METHOD

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3-Q1
LB = Q1-1.5*IQR
UB = Q3+1.5 *IQR
outliers= data[(data<(LB))| (data>(UB))]
```

```python
outliers
```
